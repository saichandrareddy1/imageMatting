```python
import keras.backend as K
from tensorflow.keras.layers import Conv2D, Concatenate, Input, Add
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.activations import relu
```

```python
# JPEG r

def JPEG_red():
    # conv(), 9x9, relu, pad
    input_tensor = Input(shape=(None, None, 3))
    x = Conv2D(32, kernel_size=(9, 9), activation='relu', padding="Valid")(input_tensor)
    # conv(), 7x7, relu, pad, 64
    x = Conv2D(64, kernel_size=(7, 7), activation='relu', padding="Valid")(x)
    # conv(), 1x1, relu, pad, 32
    x = Conv2D(32, kernel_size=(1, 1), activation='relu', padding="Valid")(x)
    # conv(), 5x5, relu, pad, 16
    x = Conv2D(16, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    # out = x
    out = Conv2D(1, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    
    model = Model(input_tensor, out)
    return model
    
def JPEG_green():
    # conv(), 9x9, relu, pad
    input_tensor = Input(shape=(None, None, 3))
    x = Conv2D(32, kernel_size=(9, 9), activation='relu', padding="Valid")(input_tensor)
    # conv(), 7x7, relu, pad, 64
    x = Conv2D(64, kernel_size=(7, 7), activation='relu', padding="Valid")(x)
    # conv(), 1x1, relu, pad, 32
    x = Conv2D(32, kernel_size=(1, 1), activation='relu', padding="Valid")(x)
    # conv(), 5x5, relu, pad, 16
    x = Conv2D(16, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    # out = x
    out = Conv2D(1, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    
    model = Model(input_tensor, out)
    return model
    
def JPEG_blue():
    # conv(), 9x9, relu, pad
    input_tensor = Input(shape=(None, None, 3))
    x = Conv2D(32, kernel_size=(9, 9), activation='relu', padding="Valid")(input_tensor)
    # conv(), 7x7, relu, pad, 64
    x = Conv2D(64, kernel_size=(7, 7), activation='relu', padding="Valid")(x)
    # conv(), 1x1, relu, pad, 32
    x = Conv2D(32, kernel_size=(1, 1), activation='relu', padding="Valid")(x)
    # conv(), 5x5, relu, pad, 16
    x = Conv2D(16, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    # out = x
    out = Conv2D(1, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    
    model = Model(input_tensor, out)
    return model
    
    
def concatination(model1, model2, model3):
    
    merge = Concatenate([model1, model2])
    final = Concatenate([merge, model3])
    
    return final
```

```python



def JPEG_alpha():
    # Jpeg_alpha # input=(27, 27, 1)
    input_tensor = Input(shape=(27, 27, 1))
    # conv(), 9x9, relu, pad
    x = Conv2D(64, kernel_size=(9, 9), activation='relu', padding='Valid')(input_tensor)
    # conv(), 7x7, relu, pad
    x = Conv2D(64, kernel_size=(7, 7), activation='relu', padding='Valid')(x)
    # feature map
    feature1= x
    # conv(), 1x1, relu
    x = Conv2D(32, kernel_size=(1, 1), activation='relu', padding='valid')(x)
    
    x = Conv2D(16, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    x = Conv2D(1, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    out = x
    
    model = Model(input_tensor, out)
    
    return model, feature1

def JPEG_beta():
    # Jpeg_alpha # input=(27, 27, 1)
    input_tensor = Input(shape=(27, 27, 1))
    # conv(), 9x9, relu, pad
    x = Conv2D(64, kernel_size=(9, 9), activation='relu', padding='Valid')(input_tensor)
    # conv(), 7x7, relu, pad
    x = Conv2D(64, kernel_size=(7, 7), activation='relu', padding='Valid')(x)
    # feature map
    feature1= x
    # conv(), 1x1, relu
    x = Conv2D(32, kernel_size=(1, 1), activation='relu', padding='valid')(x)
    
    x = Conv2D(16, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    x = Conv2D(1, kernel_size=(5, 5), activation='relu', padding="Valid")(x)
    out = x
    
    model = Model(input_tensor, out)
    
    return model, feature1

def feature_transfer(feature1, feature2):
    
    feature_transfer_add = Add()([feature1, feature2])
    
    return feature_transfer_add
    
    
def concat_add(concat, add):
    
    return Concatenate([concat, add])
```
